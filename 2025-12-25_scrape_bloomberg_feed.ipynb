```python
import feedparser
import datetime
import typing

def fetch_headlines(rss_url : str) -> typing.List[dict]:
    """
    Fetches the headlines from the specified RSS feed and returns
    them as a list of dictoinaries.
    """

    feed = feedparser.parse(rss_url)

    return feed
```
